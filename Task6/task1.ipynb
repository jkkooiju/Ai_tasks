```python
import re
file =open("c:\\Users\Hardware\Downloads\AI courses\data12.txt", "r+")
print(file.read())

#to find name
names=re.compile("\w+\s\w+")
match=names.finditer(file)

l=[]
for i in match:
    start = i.span()[0]
    end = i.span()[-1]
    l.append(file[start:end])
l
print(i)
#to find phone number
number = re.compile("[\d{11}-[+]\d{3-4}.\d[3-4]]")
where_reserch = number.finditer(file)

l=[]
for x in where_reserch:
    start = x.span()[0]
    end = x.span()[-1]
    l.append(file[start:end])
l
    
#to find email
email =re.compile("\w+\@\w+\.\w+")
where_reserch = email.finditer(file)

l=[]
for s in where_reserch:
    start = s.span()[0]
    end = s.span()[-1]
    l.append(file[start:end])
l
print(s)
```
